```python
from pymongo import MongoClient
import nltk
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer
from nltk.corpus import stopwords
import re
```

```python
def download_nltk():
    nltk.download("stopwords") #uses for stopwords
    nltk.download("punkt") # uses for tokenizing
    nltk.download("averaged_perceptron_tagger") # uses for Parts-of-Speech tagging
    
def remove_stopwords(words):
    result = []
    for w in words:
        if w not in stopwords.words("english"):
            result.append(w)
    return result

def stem_words(words):
    ps: PorterStemmer = PorterStemmer()
    result = []
    for w in words:
        result.append(ps.stem(w))
    return result

def find_keywords_in_text(text):
    text = re.sub(r"[^\w]"," ", text) # remove non-alphabetical symbols
    words = word_tokenize(text) # convert into list of words
    words = remove_stopwords(words) # remove stopwords
    words = stem_words(words) 
    keywords = nltk.pos_tag(words)
    return keywords
```

```python
# accesses database and its collection '10000 tweets'
client=MongoClient("localhost:27017")
db = client.Assignment1

download_nltk()

for t in db.Atweet.find(): 
    keywords = find_keywords_in_text(t["body"])
    for k in keywords:
        print(k)
    print("\n")
```

```python
def keywords_to_csv(keywords): 
    csv: str = "" 
    for k in keywords: 
        column: str = k[0]+","+k[1] 
        csv = csv + column + "\n" 
    return csv
```

```python
#then updates the database with the keywords in CSV format 
for t in db.Atweet.find({},{"actor":1,"body":1,"_id":0}):
    keywords=find_keywords_in_text(t["body"])
    csv=keywords_to_csv(keywords)
    db.Atweet.update_many({"actor.id":t["actor"]["id"]},{"$set":{"newkey":csv}})
```
